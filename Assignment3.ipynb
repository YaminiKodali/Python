# 4. Find the outliers per Columns-Calculate the distance- Give the justifcation

```python
#Interquartile method
import pandas as pd
import numpy as np

reader = pd.read_csv("2018-2010_export.csv")
print(sorted(reader.value))
q1, q3 = np.percentile(reader.value, [0.15,0.95])
print(q1)
#q3 = reader.value.quantile(0.95)
print(q3)
iqr = q3 - q1
print(iqr)
lower_bound = q1 - (1.5 * iqr)
print('Lower_bound', lower_bound)
upper_bound = q3 + (1.5 * iqr)
print('upper_bound', upper_bound)
print('OUTLIERS')
for i in reader.value:
 if(i > lower_bound or i < upper_bound):
  print(i)

```

```python
#z_score
import numpy as np
import pandas as pd
outliers=[]
def detect_outlier(data_1):
    
    threshold=3
    mean_1 = np.mean(data_1)
    std_1 =np.std(data_1)
    
    
    for y in data_1:
        z_score= (y - mean_1)/std_1 
        if np.abs(z_score) > threshold:
            outliers.append(y)
    return outliers
```

```python
outliers_datapoints = detect_outlier(reader.value)
print(outliers_datapoints)
print('Length :', len(outliers_datapoints))
```

# 5. Clapp the outliers

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

reader = pd.read_csv("2018-2010_export.csv")
anomalies = []
def find_anomalies(data_1):
    data_std = np.std(data_1)
    data_mean = np.mean(data_1)
    anomaly_cut_off = data_1 * 3
    
    lower_limit = data_mean - anomaly_cut_off
    upper_limit = data_mean + anomaly_cut_off
    print(upper_limit)
    print(lower_limit)
 
    for outlier in data_1 :
        if outlier > upper_limit or outlier < lower_limit:
            anomalies.append(outlier)
    return anomalies

outlier_datapoints = find_anomalies(reader.value)
clap = np.clip(outlier_datapoints)
print(clap)
```
