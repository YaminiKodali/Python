# 6. In customer table, create a new variable called “customer value segment” using customer value as follows. - 
- High Value Segment - > 25000 
- Medium Value Segment – Between 10000 and 25000 
- Low Value Segment – less than or equal to 10000 


```python
import pandas as pd

reader = pd.read_csv("Customers.csv")

CV = reader.Customer_Value
reader['Customer_value_segment'] = IF(CV > 25000,'HighSegment',IF((CV >10000) & (CV<25000),'Medium Segment','Low Segment'))
reader
 
```

# 7. Create variables “average revenue per trip” and “balance points” in the customer data set. 

```python
import pandas as pd

reader = pd.read_csv("Customers.csv")
reader["average_revenue_per_trip"] = reader.Customer_Value / reader.Buy_Times
reader["balance_points"] = reader.Points_Earned - reader.Points_Redeemed
reader
```

# 8. How many days between last purchase date and today? 

```python
import pandas as pd 
import datetime as dt

reader = pd.read_csv("Customers.csv")
today = pd.to_datetime(dt.datetime.now().date())
l_date = pd.to_datetime(reader.Recent_Date, format = '%Y%m%d')
diff = today - l_date
diff
```

# 9. Calculate percentage of sales by each last city, state and region. 

```python
import pandas as pd

reader = pd.read_csv("Customers.csv")
columns= reader[['Last_region', 'Last_state', 'Last_city', 'Customer_Value']]
percent_of_sales = columns.groupby(['Last_region', 'Last_state', 'Last_city'])['Customer_Value'].sum()/columns.Customer_Value.sum()
print('Sales Percent')
print('---------------------------------------------------------------------')
print(percent_of_sales)

```

# 10. What is the count of customers, average number of purchases and average purchase transaction value by last state and city.

```python
import pandas as pd

reader =  pd.read_csv("Customers.csv")
cust_count = reader.Customer_ID.count()
print('Customers count : ', cust_count)
avg_num_purchases = reader.Buy_Times.mean()
print('Avearge no. of purchases : ', avg_num_purchases)
cols = reader[['Last_state', 'Last_city', 'Customer_Value']]
avg_purchase_transactions = cols.groupby(['Last_state','Last_city'])['Customer_Value'].mean()
print('Average purchase transaction')
print('-----------------------------------------------------')
print(avg_purchase_transactions)


```
